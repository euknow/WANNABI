# apply, map, applymap

```python
def income_level(x):
    if x <= 39 :
        return "하"
    elif x <= 64:
        return "중"
    elif x <= 99 :
        return "상"
    else: 
        return "최상"
```

```python
data["Income"].map(income_level)
```

```python
data["Income"].map(lambda x:"하" if x <= 500 else "상")
```
